```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_point_cloud(point_cloud, ax, title="Point Cloud"):
    ax.scatter(point_cloud[:, 0], point_cloud[:, 1], point_cloud[:, 2])
    ax.set_title(title)
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')

# Select a few examples to visualize (you can adjust the number)
num_examples = 6
fig = plt.figure(figsize=(15, 10))

for idx in range(num_examples):
    # Retrieve the original and scaled point clouds using the index
    original_pc = original_shape_data[idx+8][1]  # Second element is the point cloud
    scaled_pc = scaled_map[idx+8]

    # Original point cloud
    ax = fig.add_subplot(2, num_examples, idx + 1, projection='3d')
    plot_point_cloud(original_pc, ax, title=f"Original {idx}")

    # Scaled point cloud
    ax = fig.add_subplot(2, num_examples, num_examples + idx + 1, projection='3d')
    plot_point_cloud(scaled_pc, ax, title=f"Scaled {idx}")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pickle

def plot_point_cloud(point_cloud, ax, title="Point Cloud"):
    ax.scatter(point_cloud[:, 0], point_cloud[:, 1], point_cloud[:, 2], s=1)  # Adjust point size as needed
    ax.set_title(title)
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')

# Load the modified shape data
with open('modified_shapes_data.pkl', 'rb') as file:
    modified_shape_data = pickle.load(file)
modified_map = {idx: pc for idx, pc, _ in modified_shape_data}

# Select a few examples to visualize (adjust the number as needed)
num_examples = 6
fig = plt.figure(figsize=(20, 15))

for idx in range(num_examples):
    index = idx + 8  # Adjust the index offset as needed
    # Retrieve the original, scaled, and modified point clouds using the index
    original_pc = original_shape_data[index][1]
    scaled_pc = scaled_map[index]
    modified_pc = modified_map[index]

    # Original point cloud
    ax = fig.add_subplot(3, num_examples, idx + 1, projection='3d')
    plot_point_cloud(original_pc, ax, title=f"Original {idx}")

    # Scaled point cloud
    ax = fig.add_subplot(3, num_examples, num_examples + idx + 1, projection='3d')
    plot_point_cloud(scaled_pc, ax, title=f"Scaled {idx}")

    # Modified point cloud
    ax = fig.add_subplot(3, num_examples, 2 * num_examples + idx + 1, projection='3d')
    plot_point_cloud(modified_pc, ax, title=f"Modified {idx}")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from ripser import ripser
from ripser import plot_dgms

# Load the shape data
with open('Shape_data/m_shapes_data.pkl', 'rb') as file:
    shape_data = pickle.load(file)

# Extracting indices, point clouds, and labels
indices, point_clouds, labels = zip(*shape_data)

# Convert to numpy arrays
point_clouds = np.array(point_clouds)
labels = np.array(labels)

# Define a function to compute and visualize persistence diagrams for specific indices
def visualize_persistence_diagrams(indices_to_visualize):
    for idx in indices_to_visualize:
        point_cloud = point_clouds[idx]
        label = labels[idx]
        
        # Compute the persistence diagrams using ripser
        diagrams = ripser(point_cloud)['dgms']
        
        # Visualize the persistence diagrams
        plt.figure(figsize=(8, 4))
        plt.subplot(121)
        plot_dgms(diagrams, show=True)
        plt.title(f"Shape {idx} (Label: {label}) Persistence Diagrams")
        
        plt.subplot(122)
        plt.scatter(point_cloud[:, 0], point_cloud[:, 1], c=point_cloud[:, 2], cmap='viridis')
        plt.title(f"Shape {idx} Point Cloud")
        
        plt.show()

# Specify the indices you want to visualize (e.g., [0, 1, 2] for the first three shapes)
indices_to_visualize = [0, 1, 2]  # Update this list as needed

# Call the function to visualize persistence diagrams
visualize_persistence_diagrams(indices_to_visualize)
```
