```python
import pickle

# Load the original and scaled shape data
with open('shapes_data.pkl', 'rb') as file:
    original_shape_data = pickle.load(file)

with open('scaled_shapes_data.pkl', 'rb') as file:
    scaled_shape_data = pickle.load(file)
```

```python
# Create a mapping of indices to scaled point clouds
scaled_map = {idx: pc for idx, pc, _ in scaled_shape_data}
original_map={idx: pc for idx, pc, _ in original_shape_data}
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_point_cloud(point_cloud, ax, title="Point Cloud"):
    ax.scatter(point_cloud[:, 0], point_cloud[:, 1], point_cloud[:, 2])
    ax.set_title(title)
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_point_cloud(point_cloud, ax, title="Point Cloud"):
    ax.scatter(point_cloud[:, 0], point_cloud[:, 1], point_cloud[:, 2])
    ax.set_title(title)
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')

# Select a few examples to visualize (you can adjust the number)
num_examples = 6
fig = plt.figure(figsize=(15, 10))

for idx in range(num_examples):
    # Retrieve the original and scaled point clouds using the index
    original_pc = original_shape_data[idx+8][1]  # Second element is the point cloud
    scaled_pc = scaled_map[idx+8]

    # Original point cloud
    ax = fig.add_subplot(2, num_examples, idx + 1, projection='3d')
    plot_point_cloud(original_pc, ax, title=f"Original {idx}")

    # Scaled point cloud
    ax = fig.add_subplot(2, num_examples, num_examples + idx + 1, projection='3d')
    plot_point_cloud(scaled_pc, ax, title=f"Scaled {idx}")

plt.tight_layout()
plt.show()
```

```python
# Select a few examples to visualize (you can adjust the number)
num_examples = 5
fig = plt.figure(figsize=(15, 10))

for idx, point_cloud, _ in original_shape_data[:num_examples]:
    # Retrieve the corresponding scaled point cloud
    scaled_pc = scaled_map[idx]

    # Original point cloud
    ax = fig.add_subplot(2, num_examples, idx + 1, projection='3d')
    plot_point_cloud(point_cloud, ax, title=f"Original {idx}")

    # Scaled point cloud
    ax = fig.add_subplot(2, num_examples, num_examples + idx + 1, projection='3d')
    plot_point_cloud(scaled_pc, ax, title=f"Scaled {idx}")

plt.tight_layout()
plt.show()
```
